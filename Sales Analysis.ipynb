# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#Reading the data
sales=pd.read_csv('sales.csv')
products=pd.read_csv('products.csv')
customers=pd.read_csv('customers.csv')
```

```python
#Check for any missing values in each dataset.
customers.isnull().sum()
sales.isnull().sum()
products.isnull().sum()
```

```python
# Example: Drop rows with missing values in sales_df
sales = sales.dropna()

# Example: Fill missing values in customers_df with a default value
customers['CustomerRegion'] = customers['CustomerRegion'].fillna('Unknown')
```

```python
# Setting the date as datetime format
sales['SaleDate']=pd.to_datetime(sales['SaleDate'])
```

```python
sales.dtypes
```

```python
# Example: Convert 'QuantitySold' to integer
sales['QuantitySold'] = sales['QuantitySold'].astype(int)
```

```python
sales
```

```python
#Check the duplicated rows

sales.duplicated().sum()
customers.duplicated().sum()
products.duplicated().sum()
```

```python
# Remove duplicates

sales= sales.drop_duplicates()
customers = customers.drop_duplicates()
products= products.drop_duplicates()
```

```python

# Histogram with density plot
plt.figure(figsize=(10, 6))
sns.displot(sales['TotalSales'], kde=True, color='skyblue')  # seaborn for color and density plot
plt.title('Distribution of Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Density')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')  # Adjust gridlines
plt.show()

```

```python
# Set the style and color palette
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

# Create vertical boxplot
ax = sns.boxplot(y=sales['TotalSales'], color='skyblue', fliersize=5, linewidth=1.5)

# Improve the aesthetics
plt.title('Distribution of Total Sales', fontsize=16, weight='bold')
plt.xlabel('Total Sales', fontsize=14, weight='bold')
plt.ylabel('Sales Amount', fontsize=14, weight='bold')
plt.grid(True, linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

# Data Analysis

```python
# Assuming 'SaleDate' is already in datetime format creating a new Month column
sales['Month'] = sales['SaleDate'].dt.to_period('M')
```

```python
sales
```

```python
monthly_sales=sales.groupby('Month')['TotalSales'].sum()
# Group by Month and calculate total sales
# monthly_sales = sales_df.groupby('Month')['TotalSales'].sum()
```

```python
monthly_sales
```

```python
# Plot the monthly sales trend
plt.figure(figsize=(12, 6))
plt.style.use('ggplot')
monthly_sales.plot(kind='line', marker='o', color='skyblue')
plt.title('Monthly Sales Trend',font='Serif', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Sales (BDT)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
sales['Quarter']=sales['SaleDate'].dt.to_period('Q')
Quarterly_sales=sales.groupby('Quarter')['TotalSales']
```

```python
Quarterly_sales=sales.groupby('Quarter')['TotalSales'].sum()
```

```python
# Plot the quarterly sales trend
plt.figure(figsize=(12, 6))
plt.style.use('ggplot')
Quarterly_sales.plot(kind='bar', color='skyblue',edgecolor='black')
plt.title('Quarterly Sales Trend',font='Serif', fontsize=16, fontweight='bold')
plt.xlabel('Quarter', fontsize=14)
plt.ylabel('Total Sales (BDT)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Calculate Key Performance Indicators (KPIs)

#### Average Order Value (AOV):

```python
# Calculate Average Order Value
aov = sales['TotalSales'].mean()
print(f"Average Order Value (AOV): {aov:.2f} BDT")
```

#### Customer Lifetime Value (CLV):

```python
# Calculate Customer Lifetime Value (Total Sales / Number of Customers)
total_sales_per_customer=sales.groupby('CustomerID')['TotalSales'].sum()
clv=total_sales_per_customer.mean()
print(f"Customer Life Time Value : {clv:.2f} BDT")
```

#### Sales Growth 

```python
# Calculate sales growth rate
monthly_sales_diff = monthly_sales.diff().dropna()
sales_growth_rate = (monthly_sales_diff / monthly_sales.shift(1)).dropna() * 100

# Display the average sales growth rate
average_growth_rate = sales_growth_rate.mean()
print(f"Average Monthly Sales Growth Rate: {average_growth_rate:.2f}%")
```

```python
q_dif=Quarterly_sales.diff().dropna()
quarters_growth=(q_dif/Quarterly_sales.shift(1)).dropna()*100

avg_g=quarters_growth.mean()
print(f"Average quarterly Growth rate of quarter:{avg_g:.2f} %")
```

#### Segment Customers Based on Purchase Behavior

```python
# Calculate Recency (days since last purchase)
recent_purchase = sales.groupby('CustomerID')['SaleDate'].max()   #which customer bought the products most
recency = (sales['SaleDate'].max() - recent_purchase).dt.days

# Calculate Frequency (number of purchases)
frequency = sales.groupby('CustomerID')['SaleDate'].count()

# Calculate Monetary Value (total spending)
monetary = sales.groupby('CustomerID')['TotalSales'].sum()

# Combine RFM values into a DataFrame
rfm_df = pd.DataFrame({
    'Recency': recency,
    'Frequency': frequency,
    'Monetary': monetary
})

# Display the first few rows of the RFM DataFrame
rfm_df.head(10)
```
